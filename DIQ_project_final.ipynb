#### OUTLIER ANALYSIS

Let's detect potential outliers for the 'position-related' attributes.

```python
# let's detect outliers in this distributions
def ZS(data, threshold):
    mean = np.mean(data)
    sd = np.std(data) # calculate standard deviation
    outliers = [] # create empty list to store outliers
    for i in data: # detect outliers
        z = (i - mean) / sd  # calculate z-score
        if abs(z) > threshold:  # identify outliers
            outliers.append(i)  # add to the empty list

    return outliers

# save in numpyArray the values of Codice Via, grouped by Zona
codvia_zona = DATA.groupby('Municipio')['Codice Via'].apply(list).to_numpy()
print(codvia_zona)
```

```python
# compute the outliers per Zona
outliers_by_zone = []
for i, zone in enumerate(codvia_zona):
    temp = ZS(zone, 3)
    outliers_by_zone.append(temp)
    print(f"Municipio {i+1}: {temp}")
```

```python
flat_outliers = [item for sublist in outliers_by_zone for item in sublist]
Outliers_DATA = DATA[DATA['Codice Via'].isin(flat_outliers)]
Outliers_DATA
```

We observed from the profiling step that there are a few missing values in the 'Settore_merceologico' column. For some of these, we can perform accurate imputation using the 'Insegna' column. For the remaining ones, where an accurate estimate is not possible, we will impute them using the mode of this attribute across the dataset.

```python
#Settore merceologico imputation
popular_supermarket = ['esselunga', 'pam', 'carrefour', 'supermarket', 'punto sma', 'lidl']

def fill_nan(row):
    if pd.isna(row['Settore_merceologico']):
        #check from 'Insegna'
        if row['Insegna'] in popular_supermarket:
            return 'alimentare/non alimentare'
        else:
            #impute as the mode
            return DATA['Settore_merceologico'].mode()[0]
    return row['Settore_merceologico']

# Apply the function to the DataFrame
DATA.loc[:, 'Settore_merceologico'] = DATA.apply(fill_nan, axis=1)

DATA[DATA['Settore_merceologico'].isna()].shape[0]
```

Regarding the 'Superficie...' attributes, we have integrity rules such that the sum of the first two columns must be equal to the third one. Therefore, in rows where only one value is missing out of the three, we can simply compute the missing value as the difference or sum of the other two.


```python

for index, row in DATA.iterrows():

  # Case 1: Only 'Superficie_vendita' is missing
  if np.isnan(row['Superficie_vendita[m²]']) and not np.isnan(row['Superficie_altri_usi[m²]']) and not np.isnan(row['Superficie_totale[m²]']):
    DATA.loc[index, 'Superficie_vendita[m²]'] = row['Superficie_totale[m²]'] - row['Superficie_altri_usi[m²]']

  # Case 2: Only 'Superficie_altri_usi' is missing
  elif np.isnan(row['Superficie_altri_usi[m²]']) and not np.isnan(row['Superficie_vendita[m²]']) and not np.isnan(row['Superficie_totale[m²]']):
    DATA.loc[index, 'Superficie_altri_usi[m²]'] = row['Superficie_totale[m²]'] - row['Superficie_vendita[m²]']

  # Case 3: Only 'Superficie_totale[m²]' is missing
  elif np.isnan(row['Superficie_totale[m²]']) and not np.isnan(row['Superficie_vendita[m²]']) and not np.isnan(row['Superficie_altri_usi[m²]']):
    DATA.loc[index, 'Superficie_totale[m²]'] = row['Superficie_vendita[m²]'] + row['Superficie_altri_usi[m²]']

  # Case 4: 'Superficie_vendita' and 'Superficie_altri_usi' are missing
  elif np.isnan(row['Superficie_vendita[m²]']) and np.isnan(row['Superficie_altri_usi[m²]']) and not np.isnan(row['Superficie_totale[m²]']):
    DATA.loc[index, 'Superficie_vendita[m²]'] = row['Superficie_totale[m²]']
    DATA.loc[index, 'Superficie_altri_usi[m²]'] = 0

  # Case 5: 'Superficie_totale' and 'Superficie_altri_usi' are missing
  elif not np.isnan(row['Superficie_vendita[m²]']) and np.isnan(row['Superficie_altri_usi[m²]']) and np.isnan(row['Superficie_totale[m²]']):
    DATA.loc[index, 'Superficie_totale[m²]'] = row['Superficie_vendita[m²]']
    DATA.loc[index, 'Superficie_altri_usi[m²]'] = 0

  # Case 6: Only 'Superficie_altri_usi' is non-NaN, drop the row
  elif np.isnan(row['Superficie_vendita[m²]']) and not np.isnan(row['Superficie_altri_usi[m²]']) and np.isnan(row['Superficie_totale[m²]']):
    DATA.drop(index, inplace=True)

  # Case 7: All three values are NaN, drop the row
  elif np.isnan(row['Superficie_vendita[m²]']) and np.isnan(row['Superficie_altri_usi[m²]']) and np.isnan(row['Superficie_totale[m²]']):
    DATA.drop(index, inplace=True)

  else:
    continue
```

```python
#Now that all missing values are gone i should check if all rows satisfy the consistency rules, if not correct them
CONSISTENT_ROWS = ((DATA['Superficie_totale[m²]'] > 0) &
                   (DATA['Superficie_vendita[m²]'] > 0) &
                   (DATA['Superficie_altri_usi[m²]'] >= 0) &
                   (DATA['Superficie_totale[m²]'] == (DATA['Superficie_vendita[m²]'] + DATA['Superficie_altri_usi[m²]'])))
DATA[~CONSISTENT_ROWS]
#TODO the remainin one
```
